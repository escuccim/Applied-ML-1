```python
# Load the boston house-prices data set
from sklearn.datasets import load_boston
data = load_boston()

# Fit a linear regression model
import numpy as np
x = data['data'][:, -1] # Extract LSTAT feature
y = data['target']
coefs = np.polyfit(np.log(x), y, deg=1)

# Compute prediction curve
x_values = np.linspace(min(x), max(x), num=100)
y_values = np.polyval(coefs, np.log(x_values))

# Plot predictions
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(x, y, label=None)
plt.plot(x_values, y_values, label='linear regression', color='xkcd:strawberry')
plt.title('Predicting median value of owner-occupied homes')
plt.xlabel('% lower status of the population')
plt.ylabel('Median value in $1000\'s')
plt.legend()
plt.show()
```

```python
import pandas as pd

data_df = pd.read_csv('data-points-1.csv')
data_df.head()
```

```python
x = data_df.x.values
y = data_df.y.values

print('Type of x and y:', type(x), type(y))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Plot data points
plt.scatter(x, y)
plt.show()
```

```python
import numpy as np

coefs = np.polyfit(x, y, deg=1)
print('Coefficients:', coefs) # Outputs: [ 0.77788056 -0.3904125 ]
```

```python
a, b = coefs
```

```python
x_values = np.linspace(0, 1, num=100)

# Compute the associated y values
y_values = a*x_values + b

# Plot the line
plt.scatter(x, y, label=None)
plt.plot(x_values, y_values, color='red', label='polyfit(deg=1)')
plt.legend()
plt.show()
```

```python
data_df = pd.read_csv('data-points-2.csv')
x2 = data_df.x.values
y2 = data_df.y.values
```

```python
plt.scatter(x2, y2)
plt.show()
```

```python
coefs2 = np.polyfit(x2, y2, deg=1)
print('Coefficients:', coefs2) # Outputs: [-2.25363975  1.17807746]

# Generate a hundred x values between 0 and 1
x_values2 = np.linspace(0, 1, num=100)

# Compute the associate y values
a, b = coefs2
y_values2 = a*x_values2 + b
```

```python
plt.scatter(x2, y2, label=None)
plt.plot(x_values2, y_values2, color='red', label='polyfit(deg=1)')
plt.legend()
plt.show()
```

```python
coefs2 = np.polyfit(x2, y2, deg=3)
print('Coefficients:', coefs2)
```

```python
a, b, c, d = coefs2
y_values2 = a*(x_values2**3) + b*(x_values2**2) + c*x_values2 + d
```

```python
plt.scatter(x2, y2, label=None)
plt.plot(x_values2, y_values2, color='red', label='polyfit(deg=1)')
plt.legend()
plt.show()
```

```python
y_values2 = np.polyval(coefs2, x_values2)
```
